```python
list1 = []
list2 = []
```

```python
with open("../input/input.txt", "r") as file:
    for line in file:
        numbers = line.split("   ")
        list1.append(int(numbers[0].strip()))
        list2.append(int(numbers[1].strip()))
```

```python
print(list1, list2)
```

```python
sortedList1 = sorted(list1)
sortedList2 = sorted(list2)
print(sortedList1)
```

```python
distance = 0
for i in range(len(sortedList1)):
    distance += abs(sortedList1[i] - sortedList2[i])
```

```python
distance
```

# Part-Two

```python
i = 0
j = 0

count = 0
similarityScore = 0
while i < len(sortedList1) and j < len(sortedList2):
    if sortedList1[i] < sortedList2[j]: # we've hit the limit for i
        similarityScore += count * sortedList1[i]
        i+=1
        count = 0
    elif sortedList1[i] == sortedList2[j]:
        count += 1
        j+=1
    elif sortedList1[i] > sortedList2[j]: 
        j+=1
similarityScore
```
